# Wildfire Size Class / Cause Interpretation using US Wildfire Data (SQLite)

This notebook uses the `FPA_FOD_20170508.sqlite` database (Kaggle “1.88 Million US Wildfires”)
and trains a machine learning model on the `Fires` table.

Main steps:

1. Load data from SQLite (`Fires` table)
2. Select and explore key columns
3. Feature engineering (dates, seasons, geo clusters)
4. Pre-processing and encoding
5. Train/test split and SMOTE balancing
6. XGBoost training with hyperparameter tuning
7. Model evaluation
8. Save model and encoders
9. **Enhanced inference** with:
   - Input preprocessing
   - Safe encoding
   - Area burned in acres & hectares
   - Human-readable class interpretation


Imports & Config

```python
import sqlite3
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

from imblearn.over_sampling import SMOTE
import xgboost as xgb
import joblib

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
```

Connect to SQLite & List Tables

```python
DB_PATH = "/content/drive/MyDrive/karthik/Shashank_project_DSA/wildfire-project/data/FPA_FOD_20170508.sqlite"

conn = sqlite3.connect(DB_PATH)
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
conn.close()

tables
```

Load Fires Table

```python
conn = sqlite3.connect(DB_PATH)
df_raw = pd.read_sql_query("SELECT * FROM Fires;", conn)
conn.close()

print("Raw Fires table shape:", df_raw.shape)
df_raw.head()
```

Select Relevant Columns

```python
cols = [
    "FIRE_YEAR",
    "DISCOVERY_DOY",
    "DISCOVERY_DATE",
    "LATITUDE",
    "LONGITUDE",
    "STATE",
    "FIRE_SIZE",
    "FIRE_SIZE_CLASS",
]

df = df_raw[cols].copy()
print("Selected columns:", df.columns.tolist())
print("Shape:", df.shape)
df.info()
```

Basic EDA

```python
plt.figure(figsize=(6, 4))
df["FIRE_SIZE_CLASS"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribution of FIRE_SIZE_CLASS")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10, 4))
df["FIRE_YEAR"].value_counts().sort_index().plot()
plt.title("Fires per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

Datetime Feature Engineering

```python
df["DISCOVERY_DATE"] = pd.to_datetime(df["DISCOVERY_DATE"], errors="coerce")

df["MONTH"] = df["DISCOVERY_DATE"].dt.month
df["DAY_OF_YEAR"] = df["DISCOVERY_DATE"].dt.dayofyear

df["SEASON"] = df["MONTH"].map({
    12:"Winter",1:"Winter",2:"Winter",
    3:"Spring",4:"Spring",5:"Spring",
    6:"Summer",7:"Summer",8:"Summer",
    9:"Fall",10:"Fall",11:"Fall"
})

df[["DISCOVERY_DATE", "MONTH", "DAY_OF_YEAR", "SEASON"]].head()
```

Geo Cluster Feature

```python
coords = df[["LATITUDE", "LONGITUDE"]].dropna()

kmeans = KMeans(n_clusters=10, random_state=42)
geo_labels = kmeans.fit_predict(coords)

df.loc[coords.index, "GEO_CLUSTER"] = geo_labels
df["GEO_CLUSTER"] = df["GEO_CLUSTER"].fillna(-1).astype(int)

df[["LATITUDE", "LONGITUDE", "GEO_CLUSTER"]].head()
```

Drop Unusable Columns & NaNs

```python
df = df.drop(columns=["DISCOVERY_DATE"])  # datetime not directly used by model

df = df.dropna()
print("After dropna shape:", df.shape)
df.head()
```

Encode Categorical Features

```python
encoders = {}
cat_cols = ["STATE", "SEASON", "FIRE_SIZE_CLASS"]

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le

df[cat_cols].head()
```

Define Features & Target, Split Data

```python
target = "FIRE_SIZE_CLASS"
X = df.drop(columns=[target])
y = df[target]

print("Features:", X.columns.tolist())
print("X shape:", X.shape, "y shape:", y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train:", X_train.shape, "Test:", X_test.shape)
```

Balance with SMOTE

```python
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print("Before SMOTE:", X_train.shape)
print("After SMOTE:", X_train_res.shape)
```

Train XGBoost with Hyperparameter Tuning

```python
params = {
    "learning_rate": [0.01, 0.1],
    "max_depth": [4, 6, 10],
    "n_estimators": [200, 400],
    "subsample": [0.7, 1.0],
}

xgb_model = xgb.XGBClassifier(
    objective="multi:softmax",
    num_class=len(df[target].unique()),
    tree_method="hist",
    eval_metric="mlogloss",
)

search = RandomizedSearchCV(
    xgb_model,
    param_distributions=params,
    n_iter=5,
    scoring="accuracy",
    cv=2,
    verbose=1,
    n_jobs=-1,
)

search.fit(X_train_res, y_train_res)
best_model = search.best_estimator_

print("Best parameters:", search.best_params_)
```

Evaluate Model

```python
y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification report:\n")
print(classification_report(y_test, y_pred))
```

Load Model & Enhanced Inference

```python
import joblib
import pandas as pd
import numpy as np

# -----------------------------
# Load Model From Google Drive
# -----------------------------
MODEL_PATH = "/content/drive/MyDrive/karthik/Shashank_project_DSA/wildfire-project/app/model/xgb_model.pkl"
ENC_PATH = "/content/drive/MyDrive/karthik/Shashank_project_DSA/wildfire-project/app/model/encoders.pkl"

print("Loading model and encoders...")
model = joblib.load(MODEL_PATH)
encoders = joblib.load(ENC_PATH)
print("Model and encoders loaded successfully!")

# ---------------------------------------------------------
# TRUE TRAINING FEATURE ORDER
# ---------------------------------------------------------
TRUE_FEATURE_ORDER = model.get_booster().feature_names
print("\nMODEL TRAINED FEATURE ORDER:", TRUE_FEATURE_ORDER)

# ---------------------------------------------------------
# CAUSE LABELS
# ---------------------------------------------------------
CAUSE_LABELS = {
    0: "Lightning – Natural ignition caused by lightning strikes",
    1: "Equipment Use – Caused by machinery, vehicles, or mechanical sparks",
    2: "Smoking – Ignited due to cigarettes or smoking-related activities",
    3: "Campfire / Debris Burn – Escape from campfires or debris burning",
    4: "Arson – Intentionally set wildfire",
    5: "Children / Misc Human – Accidental human-caused ignition",
    6: "Powerline / Railroad / Other – Utility or rail sparks"
}

# ==========================
# SAFE ENCODING
# ==========================
def safe_encode(series, encoder, col_name):
    known_labels = list(encoder.classes_)
    print(f"Encoding {col_name} | Known labels: {known_labels}")

    series = series.astype(str).apply(
        lambda x: x if x in known_labels else known_labels[0]
    )
    return encoder.transform(series)

# ==========================
# PREPROCESS INPUT
# ==========================
def preprocess_input(data: dict):
    print("\n--- Preprocessing Input ---")
    print("Raw input:", data)

    df = pd.DataFrame([data])

    # Auto SEASON based on MONTH or DISCOVERY_DOY
    if "MONTH" in df:
        df["SEASON"] = df["MONTH"].map({
            12:"Winter",1:"Winter",2:"Winter",
            3:"Spring",4:"Spring",5:"Spring",
            6:"Summer",7:"Summer",8:"Summer",
            9:"Fall",10:"Fall",11:"Fall"
        })
    elif "DISCOVERY_DOY" in df:
        df["MONTH"] = np.floor(df["DISCOVERY_DOY"] / 30).clip(1, 12).astype(int)
        df["SEASON"] = df["MONTH"].map({
            12:"Winter",1:"Winter",2:"Winter",
            3:"Spring",4:"Spring",5:"Spring",
            6:"Summer",7:"Summer",8:"Summer",
            9:"Fall",10:"Fall",11:"Fall"
        })
    else:
        df["SEASON"] = "Summer"

    # GEO cluster approximation
    if "LATITUDE" in df and "LONGITUDE" in df:
        df["GEO_CLUSTER"] = ((df["LATITUDE"] + df["LONGITUDE"]) // 10).astype(int)
    else:
        df["GEO_CLUSTER"] = -1

    # Safe encoding
    for col in encoders:
        if col in df:
            df[col] = safe_encode(df[col], encoders[col], col)

    # Ensure feature order
    for col in TRUE_FEATURE_ORDER:
        if col not in df:
            df[col] = 0

    df = df[TRUE_FEATURE_ORDER]

    print("\nFINAL ORDER SENT TO MODEL:", list(df.columns))
    print(df)

    return df

# ==========================
# FINAL PREDICTION
# ==========================
def predict_fire(data: dict):
    print("\n--- Running Prediction ---")
    df = preprocess_input(data)
    class_id = int(model.predict(df)[0])

    cause = CAUSE_LABELS.get(class_id, "Unknown Cause")

    # Return area burned
    area_acres = float(data.get("FIRE_SIZE", 0))
    area_hectares = round(area_acres * 0.404686, 3)

    print("\nRaw Model Output:", class_id)
    print("Cause Meaning:", cause)
    print(f"Area Burned (Acres): {area_acres}")
    print(f"Area Burned (Hectares): {area_hectares}")

    return {
        "class_id": class_id,
        "cause": cause,
        "area_burned_acres": area_acres,
        "area_burned_hectares": area_hectares
    }
```

Test Inference

```python
print("\n=== Wildfire Inference Test ===")

sample_input = {
    "FIRE_YEAR": 2015,
    "DISCOVERY_DOY": 150,
    "LATITUDE": 40.25,
    "LONGITUDE": -120.55,
    "STATE": "CA",
    "FIRE_SIZE": 12.5,
    "MONTH": 5,
    "DAY_OF_YEAR": 150
}

output = predict_fire(sample_input)

print("\n=== FINAL PREDICTION ===")
print(output)
print("============================\n")
```
